```python
import numpy as np
import pandas as pd
data = {'id':[1001,1001,1002,1003,1001,1002,1002,1003,1001,1002,1001,1002],
       'date':['2019-01-01','2019-01-01','2019-01-01','2019-01-01','2019-01-02','2019-01-02','2019-01-02','2019-01-02','2019-01-03','2019-01-03','2019-01-04','2019-01-04'],
       'goods':['A','B','A','B','A','A','B','A','A','B','A','A']}
df = pd.DataFrame(data)
df['sales'] = df.apply(lambda x:3000 if x['goods']=='A' else 2000,axis=1)
df
```

```python
pd.pivot_table(df,index='id',columns='date',values='sales',aggfunc=np.sum,fill_value=0)
```

```python
pd.pivot_table(df,index='id',columns='date',values='sales',aggfunc=np.sum,fill_value=0,margins=True)
```
