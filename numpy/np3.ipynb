```python
import numpy as np
import pandas as pd
data2 = np.random.randint(5,15,size=(5,3))
df_test = pd.DataFrame(data2,index=list('acefh'),columns=['one','two','three'],)
df_test
```

```python
df_change1 = df_change.dropna()
df_change1
```

```python
df_change2 = df_change.dropna(how='all')
df_change2
```
