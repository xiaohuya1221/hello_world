```python
import pandas as pd
import numpy as np
data = np.array([['杭州银行','600926',449,8.31],
               ['青农银行','002958',371,15.36],
               ['常熟银行','601128',237,16.01],
               ['工商银行','601398',21313,7.16],
               ['上海银行','601229',1369,7.59],
               ['江苏银行','600919',823,6.3]])
df_company = pd.DataFrame(data,columns=['公司','股票代码','市值（亿）','市盈率'])
df_company
```

```python
df_2000 = df_company[df_company['市值（亿）'].astype('int32')<2000]
df_2000
```

```python
df_2000.loc[df_2000.市盈率.astype('float64')<10]
```

```python
import numpy as np
import pandas as pd
dr = pd.date_range(start='2019-01-02', periods=100)
data = np.random.randn(100).cumsum()
close = data - np.min(data)
df = pd.DataFrame({"close": close}, index=dr)
ma5 = df.rolling(5).mean()
df['MA5'] = ma5
df
```
