```python
import numpy as np
import pandas as pd
data = pd.read_csv('houses.csv',encoding='gbk').drop_duplicates()
data.head()
```

```python
print('总房源数量',len(data.title.drop_duplicates()))
print('总小区数量',len(data.xiaoqu_name.drop_duplicates()))
print('总版块数量',len(data.sub_district_name.drop_duplicates()))
print('总房源数量',len(data.title.unique()))
```

```python
print('户型',data.huxing.unique())
print('装修',data.zhuangxiu.unique())
print('建筑年份',data.buildyear.unique())
```
